# Polynomial Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values # 1:-1 means all the columns except the last one
y = dataset.iloc[:, -1].values # -1 means the last column
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

# Fit the regressor object to the training set
lin_reg.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4)

X_poly = poly_reg.fit_transform(X)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```python
# Visualize the linear regression results
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')


# Single prediction
pred_value_1 = 6.5
pred_1 = lin_reg.predict([[pred_value_1]])
plt.plot([pred_value_1], pred_1, marker="o", markersize=5, color='orange')

plt.show()
```

## Visualising the Polynomial Regression results

```python
# Visualize the polynomial regression results
plt.scatter(X, y, color='purple')
plt.plot(X, lin_reg_2.predict(X_poly), color='pink')
plt.title('Truth or Bluff (Polynomial Regression)', color='green')
plt.xlabel('Position Level')
plt.ylabel('Salary')


# Single prediction
pred_value_2 = 4.5
pred_2 = lin_reg_2.predict(poly_reg.fit_transform([[pred_value_2]]))
plt.plot([pred_value_2], pred_2, marker="o", markersize=5, color='orange')

plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1) # 0.1 is the step size
X_grid = X_grid.reshape(len(X_grid), 1) # reshape the vector to a matrix
plt.scatter(X, y, color='purple') 
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color='pink') # X_grid instead of X
plt.title('Truth or Bluff (Polynomial Regression)', color='green')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Linear Regression

```python
lin_pred = lin_reg.predict([[10]])

print("really bad", lin_pred)
print('should be 1,000,000')
```

## Predicting a new result with Polynomial Regression

```python
lin_pred_2 = lin_reg_2.predict(poly_reg.fit_transform([[10]]))

print("not too bad", lin_pred_2)
print('should be 1,000,000')
```
