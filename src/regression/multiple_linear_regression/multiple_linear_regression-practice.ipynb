# Multiple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values # independent variables (features)
y = dataset.iloc[:, 4].values # dependent variable (profit)
```

```python
print(X)
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

columnTransformer = ColumnTransformer(
    [("encoder", OneHotEncoder(), [3])], remainder="passthrough"
)

X_transformed = columnTransformer.fit_transform(X)

# X = np.array(X_transformed)
X = X_transformed
```

```python
print(X)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(len(X_train), len(y_train))
print(len(X_test), len(y_test))
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
demo_table = []

for i in range(y_test.size):
    pred = int(regressor.predict([X_test[i]]))
    test_dependant = int(y_test[i])

    demo_table.append(
        [
            test_dependant,
            pred,
            abs(pred - test_dependant),
            "{:.2f}".format(abs(pred - test_dependant) / test_dependant * 100) + "%",
        ]
    )

pd.DataFrame(
    demo_table,
    columns=[
        "Test dependant",
        "Predicted value",
        "Absolute difference",
        "Percentage difference",
    ],
)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2) # 2 decimal places
# print y_pred and y_test side by side (vertically)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```
